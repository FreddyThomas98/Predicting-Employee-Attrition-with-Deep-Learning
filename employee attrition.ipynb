### PRE PROCESSING

```python
categorical_columns = employee_df.select_dtypes(include=['object']).columns
```

```python
employee_df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

for column in categorical_columns:
    employee_df[column] = label_encoder.fit_transform(employee_df[column])
```

```python
employee_df.head()
```

```python
num_cols = employee_df.select_dtypes(exclude=['object']).columns
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
employee_df[num_cols]= scaler.fit_transform(employee_df[num_cols])
```

```python
employee_df.head()
```

```python
from sklearn.model_selection import train_test_split
X=employee_df.drop("Attrition",axis=1)
```

```python
y=employee_df["Attrition"]
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2,stratify=y)
```

### MODEL TRAINING

```python
from sklearn.ensemble import RandomForestClassifier


rf_classifier = RandomForestClassifier()

rf_classifier.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score


y_pred = rf_classifier.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)

print("Accuracy:", accuracy)

print(classification_report(y_test, y_pred))
```

```python
X_train.shape
```

```python
from tensorflow.keras.models import Sequential
from keras.layers import Dense, Dropout


model = Sequential([
    Dense(1000, activation='relu', input_shape=(30,)),
    Dropout(0.2),
    Dense(500, activation='relu'),
    Dense(500, activation='relu'),
    Dense(500, activation='relu'),
    Dense(500, activation='relu'),
    Dense(500, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python

model.fit(X_train, y_train, epochs=100, batch_size=50)
```

```python

# Evaluate the model
test_loss, test_acc = model.evaluate(X_test, y_test, verbose=0)
print(f'Test Accuracy: {test_acc}')

# Get predictions
predictions = model.predict(X_test)
predicted_labels = np.round(predictions).flatten()

# Print classification report
print(classification_report(y_test, predicted_labels))
```
